```python
import numpy as np
from forest_helper import (
    forest_kfold_eval,
    forest_train,
    adaboost_kfold_eval,
    adaboost_train
)
from process_data_helper import (
    process_training_data,
    process_testing_data,
    load_data,
    write_predictions
)
```

```python
train_data = load_data("./data/training_data.txt")
X, Y = process_training_data(train_data, 'stats', add_bias=False)
```

```python
train_accs = []
test_accs = []
for i in range(0, 8):
    print(i)
    train_acc, test_acc = adaboost_kfold_eval(X,Y,folds=5,n_estimators=25*(i+1),split_criterion='gini',max_depth=3, min_samples_leaf=1)
    train_accs.append(np.average(train_acc))
    test_accs.append(np.average(test_acc))
    print(test_acc)
```

```python
import matplotlib.pyplot as plt
n_estimators = range(201)
n_estimators = n_estimators[25::25]
plt.figure()
plt.plot(n_estimators, test_accs, label='Testing accuracy')
plt.plot(n_estimators, train_accs, label='Training accuracy')
plt.xlabel('Number of Classifiers')
plt.ylabel('Classification accuracy')
plt.title('Accuracy vs. Number of Classifiers for Adaboost')
plt.legend(loc=0, shadow=True, fontsize='x-large')
plt.show()
```

```python
train_accs2 = []
test_accs2 = []
for i in range(0, 35):
    print(i)
    train_acc, test_acc = forest_kfold_eval(X,Y,folds=5,n_estimators=75,split_criterion='gini',max_depth=i+1, min_samples_leaf=1)
    train_accs2.append(np.average(train_acc))
    test_accs2.append(np.average(test_acc))
```
