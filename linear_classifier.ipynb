```python
import numpy as np
from svm_helper import (
    svm_kfold_eval
)
from process_data_helper import (
    process_training_data,
    process_testing_data,
    load_data,
    write_predictions
)
```

```python
train_data = load_data("./data/training_data.txt")
X, Y = process_training_data(train_data, 'l1', add_bias=True)
```

```python
Y = Y.reshape((Y.shape[0],))
train_acc, test_acc, clf = svm_kfold_eval(X,Y,folds=2,loss='log',penalty='l2',alpha=0.000001, max_iter=1000,tol=1e-3)
```

```python
print(train_acc)
print(test_acc)
```

```python
test_data = load_data("./data/test_data.txt")
Xtest = process_testing_data(test_data, 'l1', add_bias=True)
```

```python
write_predictions(clf, Xtest, './test_predictions.txt')
```
